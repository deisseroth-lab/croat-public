```python
import collections

from cellpose import models
import matplotlib.pyplot as plt
import numpy as np

NestedDict = lambda: collections.defaultdict(NestedDict)

z_slice = 100
y_slice = 200
data = np.load('data.npz')['data']
print('data shape', data.shape)
```

```python
MODEL_TYPES = ['cyto', 'nuclei']
CHANNELS = ['both', 0, 1]
ANISOTROPIES = [0.5, None, 2]
```

```python
results = NestedDict()

for model_type in MODEL_TYPES:
    model = models.Cellpose(gpu=True, model_type=model_type)
    for channel in CHANNELS:
        if channel == 'both' and model_type == 'nuclei':
            continue
    
        if channel == 'both':
            data_eval = np.moveaxis(data, 0, -1)
            channels = [1, 2]
        else:
            data_eval = data[channel]
            channels = [0, 0]
        
        for anisotropy in ANISOTROPIES:
            segments, _, _, _ = model.eval(data_eval, channels=channels, diameter=None, do_3D=True, 
                                           anisotropy=anisotropy)
            results[model_type][channel][anisotropy] = segments
```

```python
_, axes = plt.subplots(ncols=6, nrows=7, figsize=(18, 21))

axes[0][0].set_ylabel('data')

for idx, channel in enumerate(CHANNELS):
    axes[0][2*idx].set_title(f'chn={channel} YX @ z={z_slice}')
    axes[0][2*idx+1].set_title(f'chn={channel} ZX @ y={y_slice}')
    if isinstance(channel, int):
        cmap = 'Reds' if channel == 0 else 'Blues'
        axes[0][2*idx].imshow(data[channel][z_slice], cmap=cmap)
        axes[0][2*idx+1].imshow(data[channel][:, y_slice], cmap=cmap)
    else:
        axes[0][2*idx].imshow(data[0][z_slice], cmap='Reds')
        axes[0][2*idx+1].imshow(data[0][:, y_slice], cmap='Reds')
        axes[0][2*idx].imshow(data[1][z_slice], cmap='Blues', alpha=0.5)
        axes[0][2*idx+1].imshow(data[1][:, y_slice], cmap='Blues', alpha=0.5)

row = 0
for model_type in MODEL_TYPES:
    for anisotropy in ANISOTROPIES:
        row += 1
        axes[row][0].set_ylabel('model=%s    anisotropy=%s' % (model_type, anisotropy))
        for idx, channel in enumerate(CHANNELS):
            if channel == 'both' and model_type == 'nuclei':
                continue
            
            segments = results[model_type][channel][anisotropy]
            axes[row][2*idx].imshow(segments[z_slice])
            axes[row][2*idx+1].imshow(segments[:, y_slice])
```
